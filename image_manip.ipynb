## Load Sklearn image

```python
import matplotlib.pyplot as plt
from skimage import data

# 'coffee', 'astronaut', 'binary_blobs', 'brick', 'colorwheel', 'camera', 'checkerboard', 'clock', 'coffee', 'coins', 'grass', 'gravel', 'horse', 'logo', 'page', 'text', 'rocket'
image = data.coffee()
plt.imshow(image, cmap=plt.cm.gray)
plt.show()
```

```python
from skimage import data, io, filters

# show image
image = data.horse()
plt.imshow(image, cmap=plt.cm.gray)
plt.show()

# show contour
edges = filters.sobel(image)
io.imshow(edges)
plt.show()
```

## Read and download image from URL


```python
from io import BytesIO
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
import requests

url = 'https://www.thecourier.com.au/images/transform/v1/crop/frm/matthew.crossman/1d0f43ec-350b-4d77-8515-2235aead6a51.jpg/r0_0_3888_2592_w1200_h678_fmax.jpg'
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# img.show() # show in new tab
# show in notebook
imshow(np.asarray(img)) 

# Save image
img.save('output/saved_image.png')
```

## Show shades of gray

```python
from io import BytesIO
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
import requests

# fetch image
url = 'https://www.thecourier.com.au/images/transform/v1/crop/frm/matthew.crossman/1d0f43ec-350b-4d77-8515-2235aead6a51.jpg/r0_0_3888_2592_w1200_h678_fmax.jpg'
response = requests.get(url)

# Open image from response
img = Image.open(BytesIO(response.content))

# Convert to grayscale
gray_img = img.convert('L')

# Show in notebook
imshow(np.asarray(gray_img))
```

## Resize image

```python
from io import BytesIO
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
import requests

# fetch image
url = 'https://www.thecourier.com.au/images/transform/v1/crop/frm/matthew.crossman/1d0f43ec-350b-4d77-8515-2235aead6a51.jpg/r0_0_3888_2592_w1200_h678_fmax.jpg'
response = requests.get(url)

# Open image from response
img = Image.open(BytesIO(response.content))

resized = img.resize((100, 100), resample=1)

imshow(np.asarray(resized))
```

## Compress image

```python
from io import BytesIO
from matplotlib.pyplot import imshow
import numpy as np
import os
from PIL import Image
import requests

# fetch image
url = 'https://www.thecourier.com.au/images/transform/v1/crop/frm/matthew.crossman/1d0f43ec-350b-4d77-8515-2235aead6a51.jpg/r0_0_3888_2592_w1200_h678_fmax.jpg'
response = requests.get(url)

# Open image from response
img = Image.open(BytesIO(response.content))

# Compress image
img.save('output/compressed_image.png', optimize=True, quality=60)

print('Uncompressed: ', os.stat('output/saved_image.png').st_size)
print('Compressed: ', os.stat('output/compressed_image.png').st_size)
```
